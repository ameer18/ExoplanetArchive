```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Your data
data = {
    "planet_name": ["Proxima Centauri b", "TRAPPIST-1e", "Kepler-186f", "kepler 22b", "kepler 452b", "Gliese 667Cc"],
    "radius_earth": [1.17, 0.92, 1.11, 2.4, 1.6, 4.5]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Earth's radius
earth_radius = 1.0

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df["planet_name"], df["radius_earth"], color='blue', marker='o', label='Planets')
plt.axhline(earth_radius, color='green', linestyle='--', label='Earth')

plt.title("Radius of Planets Compared to Earth")
plt.xlabel("Planet")
plt.ylabel("Radius (Earth radii)")
plt.legend()

# Add minor tick lines for decimals
y_ticks = np.arange(0.5, 5.5, 0.5)  # Adjust the step to your preference
plt.yticks(y_ticks, [f"{x:.1f}" for x in y_ticks])

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```
