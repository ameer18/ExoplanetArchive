```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    "Discovery Method": ["Transit", "Radial Velocity", "Microlensing", "Imaging", "Other"],
    "Number of Exoplanets": [4332, 1056, 77, 68, 181],
    "Average Orbital Period (days)": [17.5, 722.6, 2295.4, 3592.3, 1429.8],
    "Average Planet Radius (Earth radii)": [1.12, 1.58, 0.89, 3.74, 2.21]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set a color palette
palette = sns.color_palette("Set2")

# Create a bubble chart
plt.figure(figsize=(10, 8))
for i, row in df.iterrows():
    bubble_size = row["Average Planet Radius (Earth radii)"] ** 2 * 150  # Adjust the scaling factor as needed
    
    plt.scatter(
        row["Number of Exoplanets"],
        row["Average Orbital Period (days)"],
        s=bubble_size,
        color=palette[i], alpha=0.7, edgecolors='black', label=row["Discovery Method"]
    )

    # Add labels with adjusted positions
    label_x = row["Number of Exoplanets"] + 20  # Adjust this value as needed
    label_y = row["Average Orbital Period (days)"] - 40  # Adjust this value as needed
    
  

# Set labels and title
plt.xlabel("Number of Exoplanets", fontsize=12)
plt.ylabel("Average Orbital Period (days)", fontsize=12)
plt.title("Exoplanet Discoveries by Method", fontsize=14)

# Customize grid
plt.grid(True, alpha=0.5)

# Set background color
plt.gca().set_facecolor('#f5f5f5')

# Add legend outside the plot
legend = plt.legend(title="Discovery Method", loc='center left', bbox_to_anchor=(1, 0.5))
for handle in legend.legend_handles:
    handle.set_sizes([40])  # Adjust the size of the legend markers

# Adjust layout
plt.tight_layout()

plt.show()
```
