```python
import pandas as pd
import matplotlib.pyplot as plt

# Your data
data = {
    "planet_name": ["Proxima Centauri b", "TRAPPIST-1e", "Kepler-186f", "kepler 22b", "kepler 452b", "Gliese 667Cc"],
    "orbital_period_days": [11.2, 6.1, 129.9, 290.0, 385.0, 28.0],
    "temperature_fahrenheit": [39.0, 37.0, 56.0, 22.22, 77.0, 39.6]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the percentage of temperature and orbital days
total_temperature = df["temperature_fahrenheit"].sum()
total_orbital_days = df["orbital_period_days"].sum()

df["temperature_percentage"] = (df["temperature_fahrenheit"] / total_temperature) * 100
df["orbital_days_percentage"] = (df["orbital_period_days"] / total_orbital_days) * 100

# Create a bar graph with adjusted colors
plt.figure(figsize=(10, 6))

bar_width = 0.4
plt.bar(df.index - bar_width/2, df["temperature_percentage"], width=bar_width, color='red', alpha=0.7, label='Temperature (%)')
plt.bar(df.index + bar_width/2, df["orbital_days_percentage"], width=bar_width, color='blue', label='Orbital Days (%)')

plt.title("Percentage of Temperature and Orbital Days of Planets")
plt.xlabel("Planet")
plt.ylabel("Percentage")
plt.xticks(df.index, df["planet_name"], rotation=45)
plt.legend()

plt.grid(axis='y')
plt.tight_layout()

# Add exact numbers
for i, row in df.iterrows():
    plt.text(i - bar_width/2, row["temperature_percentage"] + 2, f"{row['temperature_fahrenheit']:.2f}F", ha='center', color='black')
    plt.text(i + bar_width/2, row["orbital_days_percentage"] + 2, f"{row['orbital_period_days']:.2f} days", ha='center', color='black')

plt.show()
```
